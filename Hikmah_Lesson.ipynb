# Part 7.) Tables

Tables are a great way to store and organize a large amount of data. We can easily find the information we need!

---
#### **What are some things to know about tables?**
- They consists of rows (left to right) and columns (top to bottom)
- We can turn columns into arrays to do math on them

#### **How do I create tables?**
- Table().with_columns("Name of Column1", array1, "Name of Column 2", array 2)
- Each column you make must have an array within it
- All arrays must be the same length

````python
# We are going to make a simple table about ice cream
````

````python
# Start off with creating our data using arrays

flavors = make_array('vanilla', 'chocolate', 'strawberry', 'pistachio', 'cookies & cream')
prices = make_array(3.0, 3.0, 3.50, 4.5, 4.0)
amount_sold = make_array(10, 18, 11, 22, 19)

print(flavors)
print(prices)
print(amount_sold)
````

````python
# Create the ice cream table

ice_cream = Table().with_columns("flavor", flavors, "price", prices, "amount sold", amount_sold)
ice_cream
````

````python
# Let's find the biggest amount sold

max(ice_cream.column('amount sold'))
````

````python
# Let's find the most expensive price

max(ice_cream.column('price'))
````

````python
# Let's sort the table by amount sold

ice_cream.sort('amount sold', descending=True)
````
